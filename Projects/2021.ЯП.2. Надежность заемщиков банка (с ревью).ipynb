### Обработка пропусков

Найдем и заменим все пропуски типа NaN.

````python
# Найдем пропуски во всем датасете

data.isna().sum()
````

````python
# определим присутствуют ли пропуски в сболбцах days_employed и total_income в одних и тех же строках

gaps = data.fillna(value=0)
gaps = gaps.loc[(gaps['days_employed'] == 0) & (gaps['total_income'] == 0)]
print('В данных', gaps.shape[0], 'строк с пропусками одновременно в обоих столбцах - days_employed и total_income')
````

````python
# проверим соответствие пропусков типу занятости

print('Количество пропусков по типу занятости:')
data.loc[data['days_employed'].isna(), 'income_type'].value_counts()
````

Таким образом, пропуски не являются случайными, а представлены во всех основных типах занятости одновременно в двух столбцах: трудовой стаж и общий доход и составляют приблизительно 10% в каждом типе занятости. Из чего можно предположить, что клиенты банка намеренно скрыли эти данные. В качестве основной причины такого сокрытия можно предположить отсутствие официального источника дохода и, соответственно, трудового стажа. В случае госслужащих намеренное сокрытие могло быть связано с другими причинами, например, с государственной тайной (служащие спецслужб и пр.).

Часть значений в столбце days_employed являются отрицательными. Приведем их в положительное состояние.  

````python
# для этого вернем абсолютные значения чисел с помощью функции abs()
# чтобы применить его к колонке используем метод apply()

data['days_employed'] = data['days_employed'].apply(abs) 

# проверим минимальное значение

print('Минимальный трудовой стаж:', data['days_employed'].min())
````

Приведем данные столбца days_employed (трудовой стаж) к единой временной еденице - дню.  

````python
# Для этого все значения, которые больше максимального значения - 401755 часов = 16 739 дней (возьмем чуть больше - 16750) разделим на 24 часа 

data.loc[data['days_employed'] > 16750, 'days_employed'] = data.loc[data['days_employed'] > 16750, 'days_employed'] / 24

# проверим максимальное значение псосле операции
print('Максимальный трудовой стаж:', data['days_employed'].max(), 'дней')
````

Для столбцов days_employed и total_income (общий доход) пропуски будут заменены релевантными значениями. В качестве среднего значения в выборке для значений, которые заместят пропуски лучше подойдет медиана - это позволит избежать искажения данных из-за выбросов. Для увеличения релевантности медиана будет рассчитана и ее значение заменит пропуски в каждом типе занятости.

````python
print('Количество пропусков в столбце days_employed до обработки:', data['days_employed'].isna().sum())
print('Медианный трудовой стаж', data['days_employed'].median())

# рассчитаем медиану столбца days_employed для каждого типа занятости и сохраним в таблице med_days_employed

median_employed = data.groupby(['income_type']).agg({'days_employed': 'median'}).rename(columns = {'days_employed':'med_days_employed'})

# заменим пустные строки в столбце days_employed на медианные значения для каждого типа занятости 

data = data.merge(median_employed, on = ['income_type'])
data.loc[data['days_employed'].isna(), 'days_employed'] = data.loc[data['days_employed'].isna(), 'med_days_employed']

# проверим результат

print('Количество пропусков в столбце days_employed после обработки:', data['days_employed'].isna().sum()) 
print('Медианный трудовой стаж', data['days_employed'].median())
````

### Лемматизация

Посмотрим уникальные значения в столбце purpose (цель получения кредита)

````python
data['purpose'].unique()
````

В списке уникальных значений визуально можно выделить пять основных категории целей: недвижимость, автомобиль, образование, свадьба, ремонт. Каждая категория целей имеет несколько дубликатов. Проверим эти категории, используя метод лемматизации, т.к. в сравнени с другим методом - стеммингом - это более продвинутый алгоритм для поиска слов, записанных в разных формах.

````python
# импортируем модуль Mystem из библиотеки pymystem3

from pymystem3 import Mystem

# сохраняем функцию Mystem() в переменной m

m = Mystem() 

# склеим все значения столбца purpose в строку

purpose_lemmas = ' '.join(data['purpose'])

# лемматизируем строку в перменной purpose_lemmas и сохраним результат (список лемм) в переменную lemmas

lemmas = m.lemmatize(purpose_lemmas)

# импортируем контейнер Counter из модуля collections

from collections import Counter

# подсчитаем количество лемм 

Counter(lemmas)
````

### Категоризация данных

**Категоризация по целям взятия кредита**

Создадим новый столбец purpose_category с категориями, соответствующими каждой цели в столбце purpose.

````python
# для этого создадим функцию purpose_category

'''
Функция purpose_category возвращает название категории для каждой цели.
В качестве аргумента принимает строку.
'''

def purpose_category(row):
    lemmas = m.lemmatize(row['purpose'])
    for row in lemmas:
        if 'ремонт' in lemmas:
            return 'ремонт'
        if ('недвижимость' in lemmas or 
            'жилье' in lemmas):
            return 'недвижимость'
        if 'автомобиль' in lemmas:
            return 'автомобиль'
        if 'образование' in lemmas:
            return 'образование'
        if 'свадьба' in lemmas:
            return 'свадьба'
    return 'другая категория'


# для применения функции purpose_category к столбцу используем функцию apply() c аргументом axis=1 (применить к строкам)

data['purpose_category'] = data.apply(purpose_category, axis=1)

# подсчитаем значения в столбце
print('Количество категорий в столбце purpose_category:')
print(data['purpose_category'].value_counts())
data.head(3)
````

**Категоризация по ежемесячному доходу**  

Для начала посмотрим статистические данные по столбцу total_income (ежемесячный доход).

````python
# используем метод format(), чтобы изменить количество знаков после точки для удобства восприятия данных

data['total_income'].describe().apply("{:.2f}".format)
````

Итак, средний доход: 165.320, среднее отклонение значений в нашей выборке 98.189.  
Исходя из этих статистических показателей (округлив их значения), категоризируем уровень дохода на 4 категории по уровню дохода:

доход меньше 67.000 рублей — категория «доход ниже среднего»;  
доход от 67.000 до 143.000 рублей — категория «средний доход»;  
доход от 143.000 до 263.000 — категория «доход выше среднего»;  
доход выше 263.000 рублей — категория «высокий доход».  
  
Создадим столбец income_category с названиями данных категорий, каждая из которых соответствует доходу в столбце total_income.

````python
# для этого создадим функцию income_category

'''
Функция income_category возвращает название категории для ежемесячного дохода в стобце total_income.
В качестве аргумента принимает строку из столбца total_income.
'''

def income_category(row):
        income = row['total_income']
        if income <= 67000:
            return 'доход ниже среднего'
        if income > 67000 and income <= 143000:
            return 'средний доход'
        if income > 143001 and income <= 263000:
            return 'доход выше среднего'
        if income > 263001:
            return 'высокий доход'

data['income_category'] = data.apply(income_category, axis = 1)
print(data['income_category'].value_counts())
data.head(3)
````

**Категоризация по количеству детей**  

Категоризируем клиентов по количеству детей:  
- "без детей";  
- "1-2 ребенка";  
- "более 3 детей"  
  
Создаадим новый столбец children_category с данными категориями.

````python
# для этого создадим функцию children_category

'''
Функция income_category возвращает название категории для ежемесячного дохода в стобце total_income.
В качестве аргумента принимает строку из столбца total_income.
'''

def children_category(row):
        child = row['children']
        if child == 0:
            return 'без детей'
        if child >= 1 and child <= 2:
            return '1-2 ребенка'
        if child >= 3:
            return 'более 3 детей'

data['children_category'] = data.apply(children_category, axis = 1)
print(data['children_category'].value_counts())
data.head(3)
````

### Есть ли зависимость между наличием детей и возвратом кредита в срок?

````python
# Для сравнения наблюдений будем использовать метод свободных таблиц pivot_table().
# Аргументы метода:
    # index — столбец или столбцы, по которым группируют данные 
    # values — значения, по которым мы хотим увидеть сводную таблицу 
    # aggfunc — функция, применяемая к значениям
    
# Создадим таблицу children_pivot.    
children_pivot = data.pivot_table(index = 'children_category', 
                                          values = 'debt', # количество задолженностей по категории
                                          aggfunc = ['count','sum']) 

# Создадим в таблице children_pivot новый столбец 'ratio_debt'. 
# Cохраним в нём процент невернувших кредит от всех заемщиков каждой категории.
# Чем больше значение в столбце 'ratio_debt', тем более неблагонадежна категория заемщиков. 

# Для лучшего представления данных в столбце ratio_debt у множим результат деления на 100, округлим с помощью функции round()
# Далее используем функцию astype() для форматирования типа данных float в строковый тип 

children_pivot['ratio_debt'] = (children_pivot['sum'] / children_pivot['count'] * 100).round(1).astype('str') +'%'

# Выведем таблицу

children_pivot
````
