```python
# –¥–ª—è —ç—Ç–æ–≥–æ —Å–æ–∑–¥–∞–¥–∏–º —Å–≤–æ–¥–Ω—ã–µ —Ç–∞–±–ª–∏—Ü—ã –ø–æ –≥–æ–¥–∞–º —Å–æ —Å—Ä–µ–∑–∞–º–∏ —Å –ø—Ä–æ–ø—É—Å–∫–∞–º–∏, –ø–æ–¥—Å—á–∏—Ç–∞–≤ –∏—Ö –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤ –∫–∞–∂–¥–æ–π –∫–æ–ª–æ–Ω—Ü–µ

crit_nan = (df.loc[df['critic_score'].isna()==True]
                .pivot_table(index='year_of_release', values='name', aggfunc='count')
                .reset_index()
                     )
users_nan = (df.loc[df['user_score'].isna()==True]
                .pivot_table(index='year_of_release', values='name', aggfunc='count')
                .reset_index()
                     )

esrb_nan = (df.loc[df['rating'].isna()==True]
                .pivot_table(index='year_of_release', values='name', aggfunc='count')
                .reset_index()
                     )

# –∏ —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É –ø–æ –≥–æ–¥–∞–º –∏ –æ–±—â–µ–º—É –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –∏–≥—Ä
not_nan = (df.pivot_table(index='year_of_release', values='name', aggfunc='count')
             .reset_index()
                     )
```

#### –ñ–∞–Ω—Ä—ã –Ω–∞ –∞–∫—Ç—É–∞–ª—å–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö

–ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –æ–±—â–µ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏–≥—Ä –ø–æ –∂–∞–Ω—Ä–∞–º –Ω–∞ –∞–∫—Ç—É–∞–ª—å–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö, –≤–∫–ª—é—á–∞—è PS4.   
  
–î–ª—è —ç—Ç–æ–≥–æ –ø–æ—Å—Ç—Ä–æ–∏–º —Å—Ç–æ–ª–±—á–∞—Ç—É—é –¥–∏–∞–≥—Ä–∞–º–º—É –ø–æ –º–µ–¥–∏–∞–Ω–Ω—ã–º –ø—Ä–æ–¥–∞–∂–∞–º (–º–∏–ª–ª–∏–æ–Ω—ã –ø—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∫–æ–ø–∏–π).

```python
genre_sales = actual_platforms.pivot_table(index='genre', 
                                     values='total_sales', 
                                     aggfunc='median').sort_values('total_sales', ascending=False)
# –ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫
plt.figure(figsize=(15,7))
sns.barplot(y=genre_sales.index, x=genre_sales['total_sales'], palette='Set3')
plt.title('–†–∞–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –º–µ–¥–∏–∞–Ω–Ω—ã—Ö –ø—Ä–æ–¥–∞–∂ –ø–æ –∂–∞–Ω—Ä–∞–º', fontsize=16)
plt.ylabel('–ñ–∞–Ω—Ä—ã', size=14)
plt.xlabel('–º–µ–¥–∏–∞–Ω–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏', size=14)
plt.show()

genre_sales
```

–õ–∏–¥–µ—Ä—ã —Å—Ä–µ–¥–∏ –∂–∞–Ω—Ä–æ–≤ –ø–æ —Å—É–º–º–∞—Ä–Ω—ã–º –ø—Ä–æ–¥–∞–∂–∞–º:  
1) Shooter.  
2) Platform.  
3) Sports.  

```python
# —Å–ø–∏—Å–æ–∫ —Å —Ç–æ–ø-–∂–∞–Ω—Ä–∞–º–∏, –ø–æ –∫–æ—Ç–æ—Ä–æ–º—É –±—É–¥–µ–º —Ñ–∏–ª—å—Ç—Ä–æ–≤–∞—Ç—å –¥–∞–Ω–Ω—ã–µ –¥–ª—è —Å–≤–æ–¥–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã
top_genres = ['Shooter', 'Platform', 'Sports']

# —Å–≤–æ–¥–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ –ø–æ –∞–∫—Ç—É–∞–ª—å–Ω—ã–º –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞–º —Å –≥–ª–æ–±–∞–ª—å–Ω—ã–º–∏ –ø—Ä–æ–¥–∞–∂–∞–º–∏ –ø–æ—Å–ª–µ 2013 –≥–æ–¥–∞
top_genres_pivot = (actual_platforms
                        .loc[actual_platforms['genre'].isin(top_genres)]
                        .pivot_table(index=['genre', 'year_of_release'], values='total_sales', aggfunc='median')
                        .reset_index()
                     )

# –±–æ–∫—Å–ø–ª–æ—Ç –ø–æ —Ç–∞–±–ª–∏—Ü–µ
# —Å–Ω–∞—á–∞–ª–∞ –∏–¥–µ—Ç fig, –∞ –ø–æ—Ç–æ–º plot_top 
fig = plt.subplots(figsize=(7,7))
plot_top = sns.boxplot(x='genre', y='total_sales', data=top_genres_pivot, palette='Set3', showfliers=False)
plt.title("–ú–µ–¥–∏–∞–Ω–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä —Ç–æ–ø-–∂–∞–Ω—Ä–æ–≤ –Ω–∞ –∞–∫—Ç—É–∞–ª—å–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö", fontsize=15)
plt.xlabel('–∂–∞–Ω—Ä—ã', fontsize=12)
plt.ylabel('–º–∏–ª–ª–∏–æ–Ω—ã –ø—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∫–æ–ø–∏–π', fontsize=12)
plt.show()
```

```python
shooter = top_genres_pivot.query('genre == "Shooter"')
platform = top_genres_pivot.query('genre == "Platform"')
sports = top_genres_pivot.query('genre == "Sports"')

try:
    fig = plt.subplots(figsize = (12,5))
    plt.xlabel('–º–µ–¥–∞–Ω–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏')
    plt.ylabel('–ø–ª–æ—Ç–Ω–æ—Å—Ç—å —á–∞—Å—Ç–æ—Ç—ã')
    plt.title('–ú–µ–¥–∏–∞–Ω–Ω—ã–µ –ø—Ä–æ–¥–∞–∂–∏ –∏–≥—Ä —Ç–æ–ø-–∂–∞–Ω—Ä–æ–≤ –Ω–∞ –∞–∫—Ç—É–∞–ª—å–Ω—ã—Ö –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö')
    sns.kdeplot(shooter['total_sales'], shade=True, label="Shooter")
    sns.kdeplot(platform['total_sales'], shade=True, label="Platform")
    sns.kdeplot(sports['total_sales'], shade=True, label="Sports")
    plt.legend()
except:
    print('–∞–ø—à–∏–ø–∫–∞')
```
