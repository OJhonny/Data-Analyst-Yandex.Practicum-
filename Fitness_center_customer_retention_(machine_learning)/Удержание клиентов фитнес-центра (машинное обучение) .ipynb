# Проект "Удержание клиентов фитнес-центра"

**Проблема:**

Сеть фитнес-центров «Культурист-датасаентист» разрабатывает стратегию удержания клиентов. Для этого множество клиентских анкет были переведены в электронный вид. Необходим их анализ для построения плана по удержанию клиентов. 

**Цель:**

Провести анализ анкет клиентов фитнес-центра и подготовить рекомендации по их удержанию.

**Задачи:**
- научиться прогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;
- сформировать типичные портреты клиентов: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;
- проанализировать основные признаки, наиболее сильно влияющие на отток;
- сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами:
  - а) выделить целевые группы клиентов;
  - б) предложить меры по снижению оттока;
  - в) определить другие особенности взаимодействия с клиентами.

**План**

- Часть 1. Подготовка данных. 
- Часть 2. Исследовательский анализ данных (EDA).
- Часть 3. Построение моделей прогнозирования оттока клиентов.
- Часть 4. Кластеризация клиентов.
- Часть 5. Общие выводы и рекомендации.

## Построение моделей прогнозирования оттока клиентов

### Постановка задачи

Задача: создать модели для прогнозирования оттока клиентов и оценить их качество.

Прогнозируемая величина (целевая переменная): факт оттока клиента в следующем месяце.

Тип задач машинного обучения: обучение с учителем, бинарная классификация.

Алгоритмы моделей: логистическая регрессия и случайный лес. 

Методы оценки качества модели: Accuracy, Precision, Recall, F1-мера, ROC_AUC.

### Выделение обучающей и валидационной выборок

Для того чтобы признаки с высоким уровнем корреляции не привели к эффекту мультиколлинеарности при обучении, удалим из модели колонки month_to_end и avg_class_month.

```python
data = data.drop(['month_to_end', 'avg_class_month'], axis=1)
```

Разделим выборку на обучащую и валидационную в соотношении 80 к 20, то есть на 80% будем обучать модель, а на 20% - проверять.

```python
# разделим данные на признаки (матрица x) и целевую переменную (y)
x = data.drop('churn', axis = 1)
y = data['churn']

# разделим выборку на обучающую и валидационную в соотношении 80:20
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0, stratify=y)

# стандартизируем данные
scaler = StandardScaler()
scaler.fit(x_train)
x_train_st = scaler.transform(x_train)
x_test_st = scaler.transform(x_test)
```

### Обучение моделей

#### Логистическая регрессия

Создадим модель и обучим её на train-выборке алгоритмом логистической регрессии.

```python
# зададим алгоритм для модели
logreg_model = LogisticRegression(random_state=0)

# обучим модель
logreg_model.fit(x_train_st, y_train)

# сделаем прогнозы
logreg_predictions = logreg_model.predict(x_test_st)

# получим вероятность принадлежности объекта к первому или второму классу
logreg_probabilities = logreg_model.predict_proba(x_test_st)[:,1]
```

#### Случайный лес

Создадим модель и обучим её на train-выборке алгоритмом случайного леса.

```python
randfor_model = RandomForestClassifier(random_state=0)

randfor_model.fit(x_train_st, y_train)

randfor_predictions = randfor_model.predict(x_test_st)

randfor_probabilities = randfor_model.predict_proba(x_test_st)[:,1]
```

### Оценка качества моделей

Будем использовать следующие метрики оценки качества моделей классификации: 
- accuracy (доля правильных ответов), 
- precision (точность), 
- recall (полнота), 
- F1_score (учет баланс между precision и recall),
- ROC_AUC (площадь под кривой ошибок).

```python
# определим функцию для расчета и вывода всех метрик
def print_metrics(y_true, y_pred, y_proba, model_name):
    print("Оценка качества модели "+model_name)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
    print('\tF1: {:.2f}'.format(f1_score(y_true, y_pred)))
    print('\tROC_AUC: {:.2f}'.format(roc_auc_score(y_true, y_proba)))
```

```python
# рассчитаем метрики для модели логистической регрессии
print_metrics(y_test, logreg_predictions, logreg_probabilities, 'логистической регрессии')
print()

# рассчитаем метрики для модели случайного леса
print_metrics(y_test, randfor_predictions, randfor_probabilities, 'случайного леса')
```

Лучшие результаты показала модель логистической регрессии по всем метрикам.

2. Оценка качества моделей была проведена с использованием метрик: Accuracy, Precision, Recall, F1-мера, ROC_AUC.
2. Лучшие результаты показала модель логистической регрессии по всем метрикам.

### Обучение модели кластеризации и сегментация клиентов

Обучим модель кластеризации на основании алгоритма K-Means и сегментируем клиентов на кластеры. 

За число кластеров принимаем n=5, чтобы наши результаты можно было сравнивать с результатами остальных студентов Я.Практикума. 

```python
# зададим модель k_means с числом кластеров 5
km = KMeans(n_clusters = 5, random_state=0)

# применяем алгоритм к данным и формируем вектор кластеров
labels = km.fit_predict(x_st)

# сохраним метки кластера в датасет data
data['cluster_km'] = labels

# переименуем названия кластеров
data.loc[data['cluster_km']==0, 'cluster_km'] = 'cluster_1'
data.loc[data['cluster_km']==1, 'cluster_km'] = 'cluster_2'
data.loc[data['cluster_km']==2, 'cluster_km'] = 'cluster_3'
data.loc[data['cluster_km']==3, 'cluster_km'] = 'cluster_4'
data.loc[data['cluster_km']==4, 'cluster_km'] = 'cluster_5'

# выведем средние значения признаков для кластеров
data.groupby('cluster_km').mean()
```

Кластеры явно отличаются друг от друга по разным признакам. Но для четкого понимания различий требуется построить распределения для каждого признака по кластерам.
